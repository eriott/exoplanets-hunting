Нормируем значения flux

```python
from sklearn.preprocessing import MinMaxScaler
scaled_features = []
scaled_features = MinMaxScaler().fit_transform(X.values)
# for row in X.values:
#     scaled_features.append(MinMaxScaler().fit_transform(row.reshape(-1,1)))
    
scaled_features
```

```python
from sklearn.preprocessing import MinMaxScaler
scaled_features = []
# scaled_features = MinMaxScaler().fit_transform(X.values)
for row in X.values:
    row_scaled = MinMaxScaler().fit_transform(row.reshape(-1,1)).flatten()
    #print(row_scaled)
    scaled_features.append(row_scaled)
    
scaled_features
```

```python
X_scaled = pd.DataFrame(scaled_features, columns=X.columns)
X_scaled.head()
```

Распределение медиан интенсивности света

```python
X_scaled_median = pd.DataFrame(X_scaled.median(axis=1), columns=['median'])
X_scaled_median['LABEL'] = train['LABEL']

train_melt = pd.melt(frame=X_scaled_median, value_vars=['median'], id_vars=['LABEL'])

plt.figure(figsize=(12, 10))
ax = sns.violinplot(
    x='variable', 
    y='value', 
    hue='LABEL', 
    palette="muted", 
    split=True, 
    data=train_melt, 
    scale='area',
    scale_hue=False
)
```

Распределение средних арифметических интенсивности света

```python
X_scaled_mean = pd.DataFrame(X_scaled.mean(axis=1), columns=['mean'])
X_scaled_mean['LABEL'] = train['LABEL']

train_melt = pd.melt(frame=X_scaled_mean, value_vars=['mean'], id_vars=['LABEL'])

plt.figure(figsize=(12, 10))
ax = sns.violinplot(
    x='variable', 
    y='value', 
    hue='LABEL', 
    palette="muted", 
    split=True, 
    data=train_melt, 
    scale='area',
    scale_hue=False
)
```

Приведем данные к виду **| id звезды | time | flux**


```python
X_scaled.head()
```

```python
X_timeseries_scaled = pd.DataFrame(columns=['id', 'time', 'flux'])

id = 0
time = 0
for row in X_scaled[:5].itertuples():
    for col in row[1:]:
        X_timeseries_scaled = X_timeseries_scaled.append({'id': id, 'time': time, 'flux': col}, ignore_index=True)
        time += 1
    id += 1
```

```python
X_timeseries_scaled.head()
```

```python
ids = []
for i in range(0, X_scaled.shape[0]):
    ids += [i] * X_scaled.shape[1]

times = []
for i in range(0, X_scaled.shape[0]):
    times += list(range(0, X_scaled.shape[1]))
    
```

```python
len(ids), len(times)
```

```python
fluxes = X_scaled.values.flatten()

len(fluxes)
```

```python
X_timeseries_scaled = pd.DataFrame({'id': ids, 'time':times, 'flux':fluxes})
```
